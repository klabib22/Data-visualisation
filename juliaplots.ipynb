```julia
using Plots
using Statistics
using DataFrames
using CSV

```

```julia
df = CSV.read("AAPL Historical Data.csv", DataFrame)
```

```julia
plot(df[!,:Price], label = "Price")
plot!(df[!,:Open], label = "Open")
```

```julia
using StatsPlots
violin(df[!,"Open"], alpha = 0.7, label = "Open stocks distribution")
```

```julia
violin(df[!,"High"], alpha = 1, label = "High stocks distribution")
violin!(df[!,"Low"], alpha = 0.9, label = "Low stocks distribution")
```

```julia
histogram(df[!,:"High"], label = "High stocks histogram")
```

```julia
scatter(df[!,"Low"],alpha = 0.8, markerstrokewidth = 0.5, markersize = 2)
```

```julia
plot(df["High"], y=:High, kind = "box")
```
